#### Line Plot

- Create a line plot with year on the x-axis and population on y-axis for each `region`.

- We pivoted the dataframe for you in the variable `df_pivot`.

- Follow the Pro Tips for line plots.



```python
# quick data cleaning
df = indicators[['year', 'region', 'population']].copy()   # select columns of interest
df = df.query('region != "world"')  # only include 4 regions
df
```

```python
# Pivot the DataFrame for the line plot
df_pivot = df.pivot(index='year', columns='region', values='population').reset_index()
df_pivot.columns.name = None  # remove index name
df_pivot
```

```python
def add_end_labels(df, x, column_names, alpha=1, fontweight='normal'):
    for column_name in column_names:
        y = df[column_name].iloc[-1]
        label = f'  {column_name}'
        plt.text(x, y, label, va="center", alpha=alpha, fontweight=fontweight)

def clean_axes( ):
    ax = plt.gca()
    ax.spines[['left', 'top', 'right']].set_visible(False)
    ax.tick_params(axis='y', length=0)
    plt.grid(axis='y', alpha=0.5)

def add_axes_labels():
    y_ticks = plt.gca().get_yticks()
    y_labels = [f'{int(y_tick/1e9)}B' for y_tick in y_ticks]
    plt.yticks(y_ticks, y_labels)
    plt.xlabel('Year')   
```

```python
df_pivot.plot(x='year', y=['asia', 'africa', 'europe', 'americas'])
add_end_labels(df_pivot, x=2100, column_names=['asia', 'africa', 'europe', 'americas'], alpha=0.7)
plt.title('Continent Populations Over Time')
clean_axes()
add_axes_labels()
plt.legend().remove()
plt.show()
```

```python
# YOUR CODE HERE (add more cells as needed)

df_2000 = df.query('year == 2000').copy()
df_2100 = df.query('year == 2100').copy()

df_2000['pop_billions'] = df_2000.eval('population / 1e9')
df_2100['pop_billions'] = df_2100.eval('population / 1e9')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))

ax1.barh(df_2000['region'], df_2000['pop_billions'], color='skyblue', label='2000')
ax1.set_xlabel('Population (billions)')
ax1.set_title('Population by Region: 2000')
ax1.legend()

ax2.barh(df_2100['region'], df_2100['pop_billions'], color='orange', alpha=0.7, label='2100')
ax2.set_xlabel('Population (billions)')
ax2.set_title('Population by Region: 2100')
ax2.legend()

plt.tight_layout()
plt.show()
```
